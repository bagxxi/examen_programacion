```python
import datetime

precios_venta = {
    "A": 3800,
    "B": 3000,
    "C": 2800,
    "D": 3500
}

departamentos = [
    ["A", "B", "C", "D"],
    ["A", "B", "C", "D"],
    ["A", "B", "C", "D"],
    ["A", "B", "C", "D"],
    ["A", "B", "C", "D"],
    ["A", "B", "C", "D"],
    ["A", "B", "C", "D"],
    ["A", "B", "C", "D"],
    ["A", "B", "C", "D"],
    ["A", "B", "C", "D"]
]

vendidos = [
    ["", "", "", ""],
    ["", "", "", ""],
    ["", "", "", ""],
    ["", "", "", ""],
    ["", "", "", ""],
    ["", "", "", ""],
    ["", "", "", ""],
    ["", "", "", ""],
    ["", "", "", ""],
    ["", "", "", ""]
]

compradores = []
grilla_completa = [[" " for _ in range(4)] for _ in range(10)]

ventas_por_departamento = {departamento: 0 for departamento in precios_venta}


def mostrar_menu():
    print("----- INMOBILIARIA \"CASA FELIZ\" -----")
    print("1. Comprar departamento")
    print("2. Mostrar departamentos disponibles")
    print("3. Ver listado de compradores")
    print("4. Mostrar ganancias totales")
    print("5. Salir")

def comprar_departamento():
    global ventas_por_departamento

    flag_piso = False
    flag_departamento = False

    while not flag_piso:
        piso = int(input("Ingrese el número de piso (1-10): "))
        if piso < 1 or piso > 10:
            print("Opción inválida.")
        else:
            flag_piso = True

    while not flag_departamento:
        departamento = input("Ingrese el tipo de departamento (A-D): ").upper()
        if departamento not in ["A", "B", "C", "D"]:
            print("Opción inválida.")
        elif departamento not in precios_venta:
            print("Opción inválida.")
        else:
            flag_departamento = True

    fila = piso - 1
    columna = ord(departamento) - ord("A")

    if vendidos[fila][columna] != "":
        print("El departamento no está disponible.")
        return

    flag_rut = False
    while not flag_rut:
      run = input("Ingrese el RUN del comprador: ")
      if run.isdigit():
        if len(run) <= 8:
          flag_rut = True
          continue
        else:
          print("RUT inválido, intente nuevamente.")
      else:
        print("RUT inválido, intente nuevamente.")

    vendidos[fila][columna] = run
    compradores.append(run)
    grilla_completa[fila][columna] = "V"
    ventas_por_departamento[departamento] += 1

    print("Operación realizada correctamente.")


def mostrar_departamentos_disponibles():
    print("---- Deptos Disponibles ----")
    print("+--------------------------+")
    print("|          |       TIPO    |")
    print("|   PISO   | A | B | C | D |")
    print("+--------------------------+")
    for piso, fila in zip(range(10, 0, -1), grilla_completa[::-1]):
        print("|    {}    |".format(piso), end="")
        for estado in fila:
            print(" {} |".format(estado), end="")
        print("\n+--------------------------+")

def ver_listado_compradores():
    if len(compradores) == 0:
        print("No hay compradores registrados.")
        return

    print("----- Lista de Compradores -----")
    compradores_ordenados = sorted(compradores)
    for comprador in compradores_ordenados:
        print(comprador)


def mostrar_ganancias_totales():
    contador_ventas = 0
    total_ventas = 0
    print("+--------------------------------------------+")
    print("|-------- Ganancias Totales -----------------|")
    print("|------------------------|----------|--------|")
    print("|  Tipo de departamento  | Cantidad | Total  |")
    for departamento, cantidad in ventas_por_departamento.items():
        total_ventas_departamento = cantidad * precios_venta[departamento]
        print("|------------------------|----------|--------|")
        print(f"| Tipo {departamento} {precios_venta[departamento]} UF         |    {cantidad}     |{total_ventas_departamento} UF|")
        contador_ventas += cantidad
        total_ventas += total_ventas_departamento
    print("|------------------------|----------|--------|")
    print(f"| Total                  |    {contador_ventas}     | {total_ventas} UF|")
    print("+--------------------------------------------+\n\n\n")


def salir():
    now = datetime.datetime.now()
    print("¡Gracias por usar el sistema de Inmobiliaria \"Casa Feliz\"!")
    print("Desarrollador: Gabriel Balbontín")
    print("La fecha y hora actual es:", now.strftime("%d-%m-%Y %H:%M:%S")," ¡que tenga un gran día!")

while True:
  mostrar_menu()
  opcion = input("Seleccione una opción: ")
  if opcion == "1":
    comprar_departamento()
  elif opcion == "2":
    mostrar_departamentos_disponibles()
  elif opcion == "3":
    ver_listado_compradores()
  elif opcion == "4":
    mostrar_ganancias_totales()
  elif opcion == "5":
    salir()
    break
  else:
    print("Opción inválida. Por favor, seleccione una opción del menú.")
```
